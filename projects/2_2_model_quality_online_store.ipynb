* `Маркет_актив_6_мес` — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента.
* `Маркет_актив_тек_мес` — количество маркетинговых коммуникаций в текущем месяце.
* `Длительность` — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте.

* `Популярная_категория` — самая популярная категория товаров у покупателя за последние 6 месяцев.

* `Страниц_за_визит` — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца.

### Работа с дубликатами

Проверим на наличиие явных дубликатов

```python
print("Количество явных дубликатов:")
for i, (df_name, df) in enumerate(all_dfs.items(), start=1):
    print(f"{i}. {df_name}: {df.duplicated().sum()}")
```

Проверим наличие неявных дубликатов. Посмотрим на уникальные значения категориальных признаков. Для этого напишем пару вспомогательных функций.

```python
def get_cat_cols(df: pd.DataFrame, exclude_cols: list[str] = []) -> list[str]:
    """
    Возвращает список категориальных столбцов в датафрейме, исключая указанные колонки.

    Параметры
    ----------
    - df: Датафрейм с данными
    - exclude_cols: Список колонок, которые нужно исключить из результата
    """
    return list(df.drop(exclude_cols, axis=1).select_dtypes(object).columns)
```

```python
def print_unique_cat_vals(dfs: dict[str, pd.DataFrame]) -> None:
    """
    Вывести уникальные значения категориальных признаков для каждого датафрейма.
    """
    print("Уникальные значения категориальных признаков:")
    for i, (df_name, df) in enumerate(dfs.items(), start=1):
        print(f"{i}. Датафрейм {df_name}:")
        for j, col in enumerate(get_cat_cols(df), start=1):
            print(f'\t{j}. Признак "{translator[col]}" ({col}). Уникальные значения:')
            for val in df[col].unique():
                print("\t - ", val)
```

```python
def get_num_cols(df: pd.DataFrame, exclude_cols: list[str] = []) -> list[str]:
    """
    Возвращает список числовых столбцов в датафрейме, исключая указанные колонки.

    Параметры
    ----------
    - df: Датафрейм с данными
    - exclude_cols: Список колонок, которые нужно исключить из результата
    """
    return list(df.drop(exclude_cols, axis=1).select_dtypes(np.number).columns)


def plot_hist(
    df: pd.DataFrame,
    col: str,
    ax: Optional[Axes] = None,
    cats: Optional[str] = None,
    annotate: bool = True,
) -> Axes:
    """
    Строит гистрограмму c опциональным разбиением на категории.

    Параметры
    ----------
    - df: Датафрейм с данными
    - col: Название колонки, содержащей данные для построения гистограммы
    - ax: оси, на которых нужно строить гистограмму
    - cats: Наименование опционального столбца для разделения данных
    - annotate: Добавлят подписи к оси Х и заголовок графика
    """
    ax = sns.histplot(
        df,
        x=col,
        hue=cats,
        discrete=np.array_equal(df[col], df[col].astype(int)),
        kde=True,
        kde_kws={"bw_method": 0.6},
        edgecolor="white",
        ax=ax,
    )
    ax.set_ylabel("Количество")
    if annotate:
        ax.set_title(f'Гистограама распределение признака\n"{col}"')
        ax.set_xlabel(col)
    else:
        ax.set_xlabel("")
    return ax


def plot_box(
    df: pd.DataFrame,
    col: str,
    ax: Optional[Axes] = None,
    cats: Optional[str] = None,
    annotate: bool = True,
    **kwargs,
):
    """
    Строит боксплот c опциональным разбиением на категории.

    Параметры
    ----------
    - df: Датафрейм с данными
    - col: Название колонки, содержащей данные для построения боксплота
    - ax: оси, на которых нужно строить боксплот
    - cats: Наименование опционального столбца для разделения данных
    - annotate: Добавлят подписи к оси Х и заголовок графика
    """
    ax = sns.boxplot(df, y=col, hue=cats, ax=ax, **kwargs)
    if annotate:
        ax.set_title(f'Боксплот распределение признака\n"{col}"')
        ax.set_xlabel(col)
    else:
        ax.set_xlabel("")
        ax.set_ylabel("")


def plot_numeric(
    df: pd.DataFrame,
    num_cols: list[str],
    title: str,
    cat: Optional[str] = None,
    ncols: int = 2,
) -> None:
    """
    Функция для построения гистограмм и боксплотов по числовым признакам.

    Параметры:
    ----------
    - df: DataFrame с данными.
    - num_cols: Список числовых признаков для анализа.
    - title: Заголовок для графиков.
    - cat: Опциональный столбец для разделения данных по категориям.
    - ncols: Количество столбцов для размещения графиков.
    """

    scale = 2.5
    nrows = math.ceil(len(num_cols) / ncols)
    if ncols > 1:
        fig = plt.figure(
            layout="constrained", figsize=(scale * ncols * 2, scale * nrows)
        )
    else:
        fig = plt.figure(layout="constrained", figsize=(scale * 4, scale * nrows * 2))
    fig.suptitle(title, fontsize=16)
    subfigs = fig.subfigures(ncols=ncols, nrows=nrows, squeeze=True)
    if ncols > 1:
        subfigs = cast(list[Figure], subfigs.flatten())  # type: ignore

    for i, col in enumerate(num_cols):
        try:
            data = df[col]
        except KeyError:
            data = df[col]
        if len(num_cols) > 1:
            axs = subfigs[i].subplots(1, 2)
            subfigs[i].suptitle(col)
        else:
            axs = subfigs.subplots(1, 2)
            subfigs.suptitle(col)
        plot_hist(df, col, axs[0], cat, annotate=False)
        plot_box(df, col, axs[1], cat, annotate=False, legend=False, gap=0.2)
```

```python
def plot_cats(
    df: pd.DataFrame,
    cat_cols: Optional[list[str]] = None,
    title: str = "",
    show_vals: bool = True,
    ncols: int = 2,
) -> None:
    """
    Функция для построения графиков распределения категориальных признаков.

    Параметры:
    ----------
    - df: DataFrame с данными.
    - cat_cols: Список категориальных признаков для анализа.
    - title: Заголовок для графиков.
    - ncols: Количество столбцов для размещения графиков.
    - show_vals: Флаг, показывать ли значения на графиках.
    """

    if cat_cols is None:
        cat_cols = df.select_dtypes(object).columns.to_list()
    if len(cat_cols) == 0:
        return
    scale = 3
    nrows = math.ceil(len(cat_cols) / ncols)
    fig, axs = plt.subplots(nrows, ncols, figsize=(scale * ncols * 2, scale * nrows))
    fig.suptitle(title, fontsize=16)
    fig.set_layout_engine("constrained")
    try:
        axs = cast(Sequence[Axes], axs.flatten())
    except AttributeError:
        axs = cast(Sequence[Axes], [axs])
    for i, col in enumerate(cat_cols):
        data = df[col].value_counts()
        sns.barplot(data, ax=axs[i], orient="h", color="#8ebad9")
        axs[i].set_title(col, fontsize=14)
        axs[i].set_xlabel("")
        axs[i].set_ylabel("")
        axs[i].bar_label(
            axs[i].containers[0],
            label_type="center",
            fmt=lambda x: f"{int(x)} ({x/data.sum():.0%})",
        )
    # Удалим лишние оси
    for i in range(len(cat_cols), nrows * ncols):
        axs[i].remove()
```

```python
def highest_corrs(df: pd.DataFrame, cols: Optional[list[str]] = None, max_cols:int=10):
    """
    Функция для вычисления корреляций между парами числовых признаков

    Параметры:
    ----------
    - df: DataFrame с данными.
    - cols: Список числовых признаков для анализа.
    - max_cols: Количество самых высоких значений, которые будут показаны.
    """
    corrs = dict()
    if cols is None:
        cols = list(df.columns)
    for comb in combinations(cols, 2):
        val = df[list(comb)].phik_matrix(verbose=False).values[0, 1]  # type: ignore
        key = (translator[comb[0]], translator[comb[1]])
        corrs[key] = round(val, 2)
    s = pd.DataFrame(
        data=corrs.values(), index=list(corrs.keys()), columns=["corrs"]
    ).sort_values(
        by="corrs", key=abs, ascending=False
    )  # type: ignore

    def signigicance(cor: float) -> str:
        cor = abs(cor)
        if cor < 0.25:
            return "отсутствует"
        if cor < 0.5:
            return "слабая"
        if cor < 0.75:
            return "средняя"
        return "сильная"

    s["значимость"] = s["corrs"].transform(signigicance)
    display(s.head(max_cols))
```

### Тесты статистических гипотез

Протестируем на сколько статистически значимы распределения значения признаков для разных таргета (покупательской активности)

#### Численные признаки

Во всех случаях будем тестировать аналогичные гипотезы:

**Нуль-гипотеза**: Распределение численного признака для пользователей, у которых снизилась покупательская активность не отличается от распределения этого же признака для пользователей, у которых покупательская активность не снизилась. (Для теста Стьюдента под распределением имеется ввиду его среднее значение, а для теста Манна-Уитни общее "положение" распределения).

**Альтернативная гипотеза**: Распределения численного признака в группах **различается** (двусторонний тест).

**Уровень статистической значимости**: 0.01

```python
def stat_test(
    groups: Sequence[pd.Series],
    test: Callable,
    alternative: str = "two-sided",
    alpha: float = 0.05,
) -> None:
    """
    Функция для проведения статистического теста между группами.

    Параметры:
    ----------
    - groups: Тестируемые выборки.
    - test: Функция для проведения статистического теста (например, ttest_ind или mannwhitneyu).
    - alternative: Альтернативная гипотеза для теста (по умолчанию "two-sided").
    - alpha: Уровень значимости для проверки нулевой гипотезы (по умолчанию 0.05).

    """
    result = test(*groups, alternative=alternative)
    if result.pvalue >= alpha:
        print(
            f"Данные НЕ позволяют отвергнуть нуль гипотезу (p-value = {result.pvalue:.2e})"
        )
    else:
        print(
            f"Данные позволяют отвергнуть нуль-гипотезу в пользу альтернативной гипотезы (p-value = {result.pvalue:.2e})"
        )
```

```python
num_groups = {
    cat: [
        df_full[df_full["purchase_activity"] == unique_v][cat]
        for unique_v in df_full["purchase_activity"].unique()
    ]
    for cat in get_num_cols(df_full)
}
```

```python
alpha = 0.01
for cat, data in num_groups.items():
    print(f"* Влияние категории '{translator[cat]}':")
    print("Тест стьюдента: ", end='')
    stat_test(data, ttest_ind, alpha=alpha)
    print("Тест Манна-Уитни: ", end='')
    stat_test(data, mannwhitneyu, alpha=alpha)
    print()
```

Результаты статистических тестов говорят о том, что нуль гипотезу не следует отвергать только для категорий:
* Маркетинговая активность в текущем месяце
* Количество дней после регистрации
* Выручка в прошлом и текущем месяцах

Для всех остальных численных признаков их распределения в интересующих нас группах статистически различны.

#### Категориальные признаки

Для категориальных признаков будем проводить тест хи-квадрат с использованием таблиц сопряженности.

Во всех случаях будем тестировать аналогичные гипотезы.

**Нуль-гипотеза**: Частоты значений категориального признака для пользователей с прежней и снизившейся активностью совпадают

**Альтернативная гипотеза**: Частоты значений категориального признака для пользователей с прежней и снизившейся активностью **различаются** (двусторонний тест).

**Уровень статистической значимости**: 0.01

```python
alpha = 0.01
for cat in get_cat_cols(df_full, exclude_cols=['purchase_activity']):
    print(f"* {translator[cat]}:")
    result = chi2_contingency(pd.crosstab(df_full['purchase_activity'], df_full[cat]))
    if result.pvalue >= alpha: # type: ignore
        print(
            f"Данные НЕ позволяют отвергнуть нуль гипотезу (p-value = {result.pvalue:.2e})" # type: ignore
        )
    else:
        print(
            f"Данные позволяют отвергнуть нуль-гипотезу в пользу альтернативной гипотезы (p-value = {result.pvalue:.2e})" # type: ignore
        )
    print()
```

### Общее описание подхода к выбору лучшей модели

Используемый подход реализуется, в первую очередь, в дидактических целях, чтобы на практике сравнить влияние отдельных методов.

Оптимизация моделей будет проводиться по отдельности и будет разбита на несколько этапов для каждой модели:
1. Базовой для каждой модели будет оценка, полученная с использованием `StandardScaler` для численный данных и `OneHotencoder` для категориальных признаков, а также значениями гиперпараметров по умолчанию для каждой модели.
2. Первый этап оптимизации включает подбор гиперпараметров (а также использование альтернативных скэйлеров для признаков).
3. Второй этап включает дополнительную балансировку классов различными методами.
4. Третий этап включает дополнительную (по сравнению с первым этапом) стадию добавления (при необходимости) и отбора признаков.
5. Четвертый этап включает в себя как балансировку классов, так и работу с признаками (возможно смысла в этом особого нет, но мне хотелось посмотреть есть ли синергетический эффект балансировки классов и работы с признаками).

<a class="ancor" id="18-bullet"></a>
### Метрика

Из описания условия задачи напрашивается вывод, что для нас приоритетно выявить как можно больше пользователей, у которых снизится активность, чтобы предложить персональные акции, поэтому более приоритетна для нас метрика рекол, однако бюджет на персональные предложения не безграничен и всем подряд раздавать акции не стоит, чтобы учесть это в качестве финальной метрики для подбора порогов классификации победившей модели будем использовать метрику `f_beta_score` с `beta=2`, что будет отражать тот факт, что рекол для нас в 2 раза более важен, чем точность (precision).

При этом, для сравнения моделей между собой пользоваться будем пользоваться метрикой `roc_auc`, которая позволяет оценивать "интегральное" качество моделей.

```python
scores = {"base": {}, "opt": {}, "balance": {}, "fs": {}, "balance_fs": {}}
models = cast(dict[str, dict[str, Pipeline]], deepcopy(scores))
models.pop("base")
gs_kwargs = {"scoring": "roc_auc", "cv": 5, "n_jobs": -1, "verbose": 1}
```

### Разделение данных и дизбаланс классов

Как было установлено выше, в целевом признаке наблюдается значительный дизбаланс классов. Поэтому при разбиении данных на тренировочную и тестовую выборки будем использовать стратификацию.

Кроме того, в ходе подбора гиперпараметров протестируем несколько подходов к балансировке классов, используя инструменты библиотеки `imblearn`:
1. В качестве базового подхода к балансировке будем использовать `RandomOverSampler`.
2. Оверсэмплинг минорного класса с использованием алгоритма `SMOTE`.
3. Комбинированный подход с оверсэмплингом минорного класса и последующим прорежением обоих классов `SMOTETomek`.

```python
X = df_full.drop("purchase_activity", axis=1)
y = df_full["purchase_activity"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
)
```

Введём несколько вспомогательных функций и конструкций

```python
def show_search_result(
    search: Union[GridSearchCV, RandomizedSearchCV], n_results: int = 10
) -> None:
    """
    Возвращает датафрйем с основными результатами поиска по сетке.
    """
    df = pd.DataFrame(search.cv_results_)
    report_cols = ["rank_test_score", "mean_test_score", "std_test_score"]
    report_cols.extend([col for col in df.columns if col.startswith("param_")])
    display(df[report_cols].sort_values(by="rank_test_score").head(n_results).round(4))


def calc_grid_size(grid: Union[list, dict]) -> int:
    """
    Возвращает размер сетки для поиска параметров.
    """
    size = 0
    if isinstance(grid, list):
        for sub_dict in grid:
            size += np.array([len(container) for container in sub_dict.values()]).prod()
    else:
        size += np.array([len(container) for container in grid.values()]).prod()
    return size


def all_models_results():
    """
    Сводная таблица метрик моделей
    """
    df = (
        pd.DataFrame.from_dict(scores)
        .assign(best_cv_score=lambda x: x.max(axis=1))
        .assign(best_method=lambda x: x.idxmax(axis=1))
        .round(3)
    )
    # Добавляем значение метрики на тестовых данных
    return df


def grid_search(
    pipeline: Pipeline,
    grid: Union[list, dict],
    model: Literal["svc", "knn", "lr", "tree"],
    type: Literal["opt", "balance", "fs", "balance_fs"],
) -> None:
    """
    Осуществляет поиск по сетке, выводит и сохраняет результаты.
    """
    gs = GridSearchCV(pipeline, grid, **gs_kwargs).fit(X_train, y_train)
    scores[type][model] = gs.best_score_
    models[type][model] = gs
    show_search_result(gs)


def evaluate_params(grid: GridSearchCV):
    """
    Выводит максимальное значение метрики качества для значений гиперпараметров
    """
    df = pd.DataFrame(grid.cv_results_)
    params = [col for col in df.columns if col.startswith("param_")]

    for param in params:
        display(
            df[[param, "mean_test_score", "rank_test_score"]]
            .groupby(param, sort=False)
            .agg({"mean_test_score": "max", "rank_test_score": "min"})
            .sort_values(by="mean_test_score", ascending=False)
            .head(10)
            .round(4)
        )
```

<a class="ancor" id="20-bullet"></a>
### Скалирование признаков

Скалирование (стандартизация) численных признаков особенно важно для моделей, измеряющих "расстояние" между признаками, а также для линейных моделей. "Древесные" модели в меньшей степени зависят от скалирования численных призанков.

В качестве базового метода скалирования численных признаков будем использовать `StandardScaler`, который приводит распределние численного признака к нулевому среднему и единичному стандартному отколонению. В качестве альтернативного метода скалирования численных признаков попробуем использовать `QuantileTransformer` с `output_distribution='normal'`. Этот метод сохраняет ранги численных значений, но делает само распределение численных значений близким к стандартному нормальному.

Для кодирования категориальных признаков в качестве базового метода будем использовать `OneHotEncoder` с `drop='fisrst`. В качестве альтернативного метода попробуем `OrdinalEncoder`, использование этого альтернативного энкодера может быть особенно полезным для древесных моделей, которые не любят слишком большого количества признаков (большое количество признаков может приводить к переобучению).

```python
base_scaler = ColumnTransformer(
    [
        ("num_scaler", StandardScaler(), make_column_selector(dtype_include=np.number)),
        (
            "cat_scaler",
            OneHotEncoder(drop="first", sparse_output=False),
            make_column_selector(dtype_include=object),
        ),
    ],
    verbose_feature_names_out=False,
)
```

### SVC

Гиперпараметры и диапазоны их значений для оптимизации:
1. **Ядро** Попробуем "стандратные" ядра: `linear`, `rbf`, `sigmoid` и `poly`.
2. **Регуляризация** Используем значения коэффициента регуляризации в интервале $[10^{-2},10]$.
3. **Гамма** Типичные значения скалируются как $1/{n\_samples}$, В нашем случае с учетом кросс-валидации это около 0.005, поэтому используем значения в интервале $[10^{-4},1]$.
4. Для полиномиального ядра попробуем степени от 2 до 5.

```python
svc_base_pipeline = Pipeline(
    [("scaler", base_scaler), ("svc", SVC(random_state=RANDOM_STATE, tol=0.01))]
)
scores["base"]["svc"] = cross_val_score(
    svc_base_pipeline, X_train, y_train, scoring="roc_auc", n_jobs=-1
).mean()
print(f'Базовый скор модели SVC: {scores["base"]["svc"]:.3f}')
```

Первый шаг оптимизации модели: подбор гиперпараметров

```python
scaler_params = {
    "scaler__num_scaler": [
        StandardScaler(),
        QuantileTransformer(n_quantiles=500, output_distribution="normal"),
    ],
    "scaler__cat_scaler": [
        OneHotEncoder(drop="first", sparse_output=False),
        OrdinalEncoder(),
    ],
}

# Сетки подбора гиперпарамаетров различны для разных ядер
# Сетка гиперпараметров для линейного ядра
svc_linear_param_dict = {
    "svc__kernel": ["linear"],
    "svc__C": [0.01, 0.1, 1, 10],
    **scaler_params,
}

# Сетка гиперпараметров для нелинейных (но не полноимиального) ядер
nonlinear_svc_param_dict = svc_linear_param_dict.copy()
nonlinear_svc_param_dict.update(
    {"svc__kernel": ["rbf", "sigmoid"], "svc__gamma": [0.001, 0.01, 0.1, 1]}
)

# Сетка гиперпараметров для полиномиального ядра
poly_svc_param_dict = nonlinear_svc_param_dict.copy()
poly_svc_param_dict.update({"svc__kernel": ["poly"], "svc__degree": [2, 3, 4]})

svc_params = [svc_linear_param_dict, nonlinear_svc_param_dict, poly_svc_param_dict]
grid_search(svc_base_pipeline, svc_params, "svc", "opt")
```

Перейдём ко второму этапу и добавим балансировку классов.

```python
svc_balance_pipeline = imblearn.pipeline.Pipeline(
    [
        ("scaler", base_scaler),
        (
            "balancer",
            imblearn.over_sampling.RandomOverSampler(random_state=RANDOM_STATE),
        ),
        ("svc", SVC(random_state=RANDOM_STATE, tol=0.01)),
    ]
)
balance_params = {
    "balancer": [
        imblearn.over_sampling.RandomOverSampler(random_state=RANDOM_STATE),
        imblearn.over_sampling.SMOTE(random_state=RANDOM_STATE),
        imblearn.combine.SMOTETomek(random_state=RANDOM_STATE),
    ]
}
svc_balance_nonlinearn_param_dict = {**nonlinear_svc_param_dict, **balance_params}
svc_balance_poly_param_dict = {**poly_svc_param_dict, **balance_params}
svc_balance_params = [svc_balance_nonlinearn_param_dict, svc_balance_poly_param_dict]

grid_search(svc_balance_pipeline, svc_balance_params, "svc", "balance")
```

<a class="ancor" id="26-bullet"></a>
### Подбор порогов классификации

Как уже говорилось ранее, для подбора порогов будем использовать метрику `fbeta` со значением `beta=2`.
Построим распределения метрики и найдём оптимальные пороги для всех моделей

```python
def plot_fbeta(model: Pipeline, ax: Axes, label: Optional[str] = None) -> None:
    """
    Находит оптимальное значение порога и строит распределение метрики.
    """
    try:
        probas = pd.Series(model.predict_proba(X_train)[:, 1])
    except AttributeError:
        probas = pd.Series(model.decision_function(X_train))
    thrs = probas.sort_values().unique()
    scores = np.zeros_like(thrs)
    for i, thr in enumerate(thrs):
        prediction = probas.transform(
            lambda x: model.classes_[1] if x > thr else model.classes_[0]
        )
        scores[i] = fbeta_score(
            y_train, prediction, beta=2, pos_label=model.classes_[1]
        )
    max_i = scores.argmax()
    print(
        f"\t{label + ": " if label is not None else ''}Лучшее значение метрики {scores[max_i]:.3f} достигается при пороге {thrs[max_i]:.2f}"
    )

    ax.plot(thrs, scores, linestyle="-", label=label)
```

#### Метрика
В качестве финальной метрики для подбора порогов классификации победившей модели использована метрика `f_beta_score` с `beta=2`, что отражает тот факт, что рекол для нас в 2 раза более важен, чем точность (precision).

При этом, для сравнения моделей между собой пользоваться использовалась метрика `roc_auc`, которая позволяет оценивать "интегральное" качество моделей.


#### Дизбаланс классов
В ходе подбора гиперпараметров протестировано несколько подходов к балансировке классов, используя инструменты библиотеки `imblearn`:

3. Комбинированный подход с оверсэмплингом минорного класса и последующим прорежением обоих классов `SMOTETomek`.

В качестве базового метода скалирования численных признаков использовался `StandardScaler`, который приводит распределние численного признака к нулевому среднему и единичному стандартному отколонению. В качестве альтернативного метода скалирования численных признаков использован `QuantileTransformer` с `output_distribution='normal'`. Этот метод сохраняет ранги численных значений, но делает само распределение численных значений близким к стандартному нормальному.

Для кодирования категориальных признаков в качестве базового метода использован `OneHotEncoder` с `drop='fisrst`. В качестве альтернативного метода использован `OrdinalEncoder`, использование этого альтернативного энкодера может быть особенно полезным для древесных моделей, которые не любят слишком большого количества признаков (большое количество признаков может приводить к переобучению).

## 8. Сегментация покупателей

Для начала соберем в один датафрейм инфомрацию о доходности пользователей, предсказания вероятностей снижения их покупательской активности и всё их признаковое описание.

```python
predicts = cast(pd.DataFrame, X_test.copy())
predicts["predicts"] = best_model.predict_proba(X_test.rename(translator, axis=1))[:, 1]
predicts = predicts.join(df_money, how="left")
```

```python
ax = sns.scatterplot(predicts, x="predicts", y="profit")
ax.set_title(
    "Диаграмма рассеяния прибыльности и вероятности снижения покупательской активности."
)
ax.set_xlabel(
    "'Вероятность' снижения покупательской активности \nчем выше значение - тем выше вероятность"
)
_ = ax.set_ylabel("Средняя прибыльность пользователя")
```

Зададим сегментацию пользователей:
* Во-первых разделим пользователей по вероятности снижения их покупательской активности:
  * **Высокая** категория снижения активности - пользователи с "decision_score" больше 0.9
  * **Средняя** категория снижения активности - пользователи с "decision_score" от 0.9 до 0.2
  * **Низкая** категория снижения активности - пользователи с "decision_score" ниже 0.2
  * **Нет** категория снижения активности - пользователи с "decision_score" меньше 0.05

При такой классификации наша оптимальная модель (`knn`) пометит в качестве пользователей со уменьшающейся покупательской активностью пользователей категорий **высокая** и **средняя**. Пользователи категории **Низкая** будут помечены как сохраняющие свою активность, но уверенность модели в таком предсказании не очень высока.

```python
def mark_decrease_activity(val: float) -> str:
    if val > 0.9:
        return "высокая"
    if val > 0.2:
        return "средняя"
    if val > 0.05:
        return "низкая"
    return "нет"


predicts["predict_cats"] = pd.Categorical(
    predicts["predicts"].transform(mark_decrease_activity),
    categories=["нет", "низкая", "средняя", "высокая"],
    ordered=True,
)
```

Также разобъём всех пользователей по доходности на 3 категории с равным количеством пользователей в каждой категории

```python
predicts["profit_cats"] = pd.qcut(
    predicts["profit"], 3, labels=["низкая", "средняя", "высокая"]
)
translator["predict_cats"] = "Вероятность снижения активности"
translator["profit_cats"] = "Категория прибыльности"
```

Теперь сгруппируем признаки пользователей по категориям риска снижения активности и доходности.

Для численных признаков будем вычислять медиану, а для категориальных самое частое значение.

```python
for cat in predicts.drop(
    ["profit_cats", "predict_cats", "promo_only", "profit", "predicts"], axis=1
):
    print(f"{translator[cat]}:")
    ax = sns.catplot(
        predicts.rename(translator, axis=1),
        x=translator["profit_cats"],
        y=translator[cat],
        col=translator["predict_cats"],
    )
    if pd.api.types.is_numeric_dtype(predicts[cat]):
        display(
            predicts.pivot_table(
                index="predict_cats",
                columns="profit_cats",
                values=cat,
                aggfunc="median",
                observed=False,
            )
        )
    else:
        display(
            predicts.pivot_table(
                index="predict_cats",
                columns="profit_cats",
                values=cat,
                aggfunc=lambda x: x.mode(),
                observed=False,
            )
        )
    plt.show()
```

   2. Первый этап оптимизации включает подбор гиперпараметров (а также использование альтернативных скэйлеров для признаков).
   3. Второй этап включает дополнительную балансировку классов различными методами.
   4. Третий этап включает дополнительную (по сравнению с первым этапом) стадию добавления (при необходимости) и отбора признаков.
   5. Четвертый этап включает в себя как балансировку классов, так и работу с признаками (возможно смысла в этом особого нет, но мне хотелось посмотреть есть ли синергетический эффект балансировки классов и работы с признаками).

   #### Метрика
   В качестве финальной метрики для подбора порогов классификации победившей модели использована метрика `f_beta_score` с `beta=2`, что отражает тот факт, что рекол для нас в 2 раза более важен, чем точность (precision).

   При этом, для сравнения моделей между собой пользоваться использовалась метрика `roc_auc`, которая позволяет оценивать "интегральное" качество моделей.


   #### Дизбаланс классов
   В ходе подбора гиперпараметров протестировано несколько подходов к балансировке классов, используя инструменты библиотеки `imblearn`:
   1. В качестве базового подхода к балансировке будем использовать `RandomOverSampler`.
   2. Оверсэмплинг минорного класса с использованием алгоритма `SMOTE`.
   3. Комбинированный подход с оверсэмплингом минорного класса и последующим прорежением обоих классов `SMOTETomek`.

   #### Скалирование признаков

   В качестве базового метода скалирования численных признаков использовался `StandardScaler`, который приводит распределние численного признака к нулевому среднему и единичному стандартному отколонению. В качестве альтернативного метода скалирования численных признаков использован `QuantileTransformer` с `output_distribution='normal'`. Этот метод сохраняет ранги численных значений, но делает само распределение численных значений близким к стандартному нормальному.

   Для кодирования категориальных признаков в качестве базового метода использован `OneHotEncoder` с `drop='fisrst`. В качестве альтернативного метода использован `OrdinalEncoder`, использование этого альтернативного энкодера может быть особенно полезным для древесных моделей, которые не любят слишком большого количества признаков (большое количество признаков может приводить к переобучению).


   #### Результаты подбора гиперпараметров

   * Модель `SVC` с гиперпараметрами по умолчанию показала неплохие реузльтаты сходу, для всех остальных моделей потребовалась донастройка гиперпараметров и работа с признаками, чтобы добиться повышения качества модели.
   * Балансировка классов с использованием различных методик (`RandomOverSampling`, `SMOTE`, `SMOTETomek`) сама по себе ниразу не привела к улучшению используемой метрики качества модели.
   * Отбор призанков (`FeatureSelection`) позволил дополнительно улучшить качество всех моделей по сравнению с обычным поиском по сетке.
   * Использование балансировки классов вместе с отбором признаков не дало синергетического эффекта в 3 из 4 случаев, однако логистической регрессии этот подход всё же позволил добиться максимального значения метрики качества (хотя разница по сравнению с отбором признаков без использования балансировки минимальна и явно статистически не значима). В целом, балансировка классов не показала какой-либо пользы.
   * Лучшей стратегией в большинстве случаев является поиск гиперпараметров по сетке с отбором признаков.
   * Как на тренировочных, так и на тестовых данных лучшие результаты показала модель `KNN`. Стоит отметить, что на тестовых данных метрика модели `SVC` очень близка к лидеру.

   Нужно также отметить, что в целом лидерами остались модели `SVC` и `KNN`, что близко к выводам, сделанным с использованием метрики `roc_auc`.

   4. Доля акционных покупок

       * **Средняя** категория снижения активности - пользователи с "decision_score" от 0.9 до 0.2
       * **Низкая** категория снижения активности - пользователи с "decision_score" ниже 0.2

    * По прибыльности: на 3 категории (низкая, средняя, высокая) с равным количеством пользователей в каждой категории

    Анализ распределений признаков:
    * У клиентов с высокой вероятностью снижения активности и высокой и средней прибыльностью самые низкие средние показатели маркетинговой активности за последние 6 месяцев.

    * Пользователи с высокой вероятностью снижения активности во всех категориях доходности просматривают мало страниц и категорий, и в целом проводят мало времени на сайте.
    * Пользователи с высокой вероятностью снижения активности во всех категориях доходов в среднем имеют бОльшее количество неоплаченных товаров в корзине (более 5 товаров в корзине имеют только пользователи с высокой и средней вероятностью снижения активности).
